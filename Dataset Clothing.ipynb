```python
import os
import pandas as pd
import numpy as np
import glob
from PIL import Image
import os.path
```

```python
os.chdir('C:\\Nikita\\Dataset Clothing\\dataset')
```

```python
data = pd.DataFrame()

data['Folder'] = True
data['Files'] = True
```

```python
dirs = []
files = []
for dirpath, dirnames, filenames in os.walk('.'):
    for dirname in dirnames:
        if len(dirname) != 0:
            dirs.append(dirname)
    if len(filenames) > 1:
        files.append(len(filenames))
    #print(dirnames, len(filenames))
    #data.append({'Folder': dirnames, 'Files': len(filenames)}, ignore_index = True)
```

```python
lines = zip(dirs, files)
for line in lines:
    print(line[0], line[1])
    data = data.append({'Folder': line[0], 'Files': line[1]}, ignore_index = True)
```

```python
data2 = pd.DataFrame()

data2['File'] = True
data2['Path'] = True
data2['File size'] = True
data2['Pic dimensions'] = True
```

```python
data2
```

```python
for entry in sorted(glob.glob('C:/Nikita/Dataset Clothing/dataset/**/*.jpg', recursive=True)):
    try:
        imsize = Image.open(entry).size
    except:
        imsize = np.nan
    data2 = data2.append({'File': entry[entry.rfind('\\')+1:], 'Path': entry, 'File size': os.path.getsize(entry), 'Pic dimensions': imsize}, ignore_index = True)
data2.sort_values(by='File size')
```

```python
data2.to_csv('C:/Nikita/Dataset Clothing/sizes_database.csv', header = True, index = True)
```
