```python
annotations = pd.DataFrame()
annotations['File'] = True
annotations['Smallest box dimension'] = True
annotations
```

```python
for file in glob.glob('C:/Nikita/Dataset Clothing/annotations/**/*.json', recursive=True):
    try:
        with open(file, "r") as read_file:
            jsfile = json.load(read_file)
    except:
        print('Failed to read the json file, I don\'t know why')
    smallest_box_dimension = min(jsfile['arr_boxes'][0]['width'], jsfile['arr_boxes'][0]['height'])
    for item in jsfile['arr_boxes']:
        if min(item['width'], item['height']) < smallest_box_dimension:
            smallest_box_dimension = min(item['width'], item['height'])
    annotations = annotations.append({'File': jsfile['file_name'], 'Smallest box dimension': smallest_box_dimension}, ignore_index = True)
```

```python
annotations.sort_values(by = 'Smallest box dimension')
```
